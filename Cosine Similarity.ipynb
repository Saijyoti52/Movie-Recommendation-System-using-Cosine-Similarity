```python
#Import Libraries
import numpy as np
import pandas as pd
```

```python
#import movies dataset
movies_df=pd.read_csv("movies.csv")
```

```python
#Import Ratings dataset
ratings_df=pd.read_csv("ratings.csv")
```

```python
#MERGING MOVIES and RATINGS
MOVIES_RATINGS=pd.merge(ratings_df,movies_df,on='movieId')
```

```python
Movie_RatingCount=MOVIES_RATINGS.groupby(by=['title'])['rating'].count().reset_index().rename(columns={'rating':"Ratings_Count"})
```

```python
Movie_RatingCount
```

```python
RatingswithRatingsCount=MOVIES_RATINGS.merge(Movie_RatingCount,how='outer',on='title')
#Ratings_RatingsCount=MOVIES_RATINGS.merge(Movie_RatingCount,left_on='title',right_on='title',how='left')
```

```python
RatingswithRatingsCount
```

```python
pd.set_option('display.float_format', lambda x: '%.3f' % x)
print(RatingswithRatingsCount['Ratings_Count'].describe())
```

```python
threshold=50
popular_rated_movies=RatingswithRatingsCount.query('Ratings_Count >= @threshold')
```

```python
popular_rated_movies
```

```python
popular_rated_movies.shape
```

```python
#pivot matrix
movies_userratings=popular_rated_movies.pivot_table(index='title',columns='userId',values='rating').fillna(0)
```

```python
movies_userratings
```

```python
#Converting the pivot table inot array matrix which will help to fit our model
from scipy.sparse import csr_matrix
movie_features_matrix  = csr_matrix(movies_userratings.values)
```

```python
print(movie_features_matrix)
```

```python
from sklearn.neighbors import NearestNeighbors
model_knn = NearestNeighbors(metric = 'cosine', algorithm = 'brute')
model_knn.fit(movie_features_matrix)
```

```python
#help(NearestNeighbors)
```

```python
movies_userratings.shape
```

```python
movie_features_matrix.shape
```

```python
query_index = np.random.choice(movies_userratings.shape[0])
print(query_index)
distances, indices = model_knn.kneighbors(movies_userratings.iloc[query_index].values.reshape(1, -1),n_neighbors = 6)
```

```python
print(distances,indices)
#print(type(distances))
```

```python
#movies_userratings.index[1102]
```

```python
for i in range(0, len(distances.flatten())):
    if i == 0:
        print('Recommendations for {0}:\n'.format(movies_userratings.index[query_index]))
    else:
        print('{0}: {1}, with distance of {2}:'.format(i, movies_userratings.index[indices.flatten()[i]], distances.flatten()[i]))
```
